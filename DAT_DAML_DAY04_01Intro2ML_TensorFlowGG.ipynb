### The code..

So now let's take a look. This is the entire code that we could use to create a machine-learned model that figures out what matches these numbers to each other.

p09.PNG

#### Define the model

p10.PNG

This first line defines the <font color=ff00ff>model</font> itself. A model is a <font color=ffff00>trained neural network</font>, and here we have the simplest possible neural network one, which in this case, is a single layer indicated by the ```keras.layers.Dense(...)``` code. 

This layer has a single neuron in it, indicated by ```units=1```. We also feed a single value into the neural network, which is the *X* value, and we'll have the neural network to predict what the *Y* would be for that *X*. So that's why we just say the ```input_shape``` is one value.




---



#### Compile the model

p14.PNG

When we <font color=ff00ff>compile</font> the model, there are two functions: the <font color=ffff00>loss</font> and the <font color=00ffff>optimizer</font>. These are the <font color=03f72f>key</font> to the machine learning. 

How machine learning works is that the <font color=00ff00>model</font> will <font color=ff00ff>make a guess</font> about the relationship between the numbers. For example, it might guess that **y = 5X + 5**.

And when training, it will then calculate <font color=ff00ff>*how good*</font> or <font color=ff00ff>*how bad*</font> that the guess is, using the <font color=ffff00>**loss** function</font>. And then, it will use the <font color=00ffff>**optimizer** function</font> to generate another <font color=ff00ff>guess</font>.

---


fit.png 

The logic is that the combination of those two functions (ie., loss and optimizer) will <font color=ff00ff>slowly</font> get us <font color=ff00ff>closer</font> and <font color=ff00ff>closer</font> to the correct formula. 

And, in this case, it will go through that loop 500 times, <font color=ff00ff>making a guess</font>, <font color=ffff00>calculating how accurate that guess is</font>, and the <font color=00ffff>using the *optimizer* to enhance that guess</font>, and so on. 

---

#### Provide the data and Fit the model

prog.png

The data itself is set up as an <font color=00ff00>array of Xs and Ys</font>. 

The process of training the neural network, where it learns the relationship between the X's and Y's, is in the ```model.fit()``` call. We literally say, fit the Xs to the Ys and try this 500 times. 

That's where it will go through the loop before making a guess, measuring how good or bad it is (the loss), or using the optimizer to make another guess. It will do that for the number of epochs that you specify. When you run that code, you'll see the loss will be printed out for each epoch.




---



#### Predict the new input

p19.PNG

When it was done, we will have a <font color=ff00ff>trained model</font>. So now we can try to predict a Y value for a given X.

When we try this line of code, predict the Y when X equals 10? We might think that the answer is exactly 19? 

But it isn't. It will print actually something like 18.9998! It's close to 19, but it's not quite there. Why do we think that would be?

As the computer was trained to match only six pairs of numbers, it looks like a straight-line relationship between them, for those six pairs of numbers, but it may not be a straight line for values outside of those six. There is a very <font color=00ffff>high probability</font> that it is a straight line, but we can't be certain. And this probability is built into the <font color=ffff00>prediction</font>, so it's telling us a value very close to 19, instead of exactly 19.




---



### The complete code

````
# the simplest ML code to learn the relationship of X and Y
from tensorflow import keras
import numpy as np

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])]) # 1 node 1 input 
model.compile(optimizer='sgd', loss='mean_squared_error') # loss  error  sgd 

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)  # input
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)#

model.fit(xs, ys, epochs=500) # train 500 

print(model.predict([10.0]))
print('Done..')
````

### Bonus Topic:
What the parameter ML has actually learned?

````
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
model.fit(xs, ys, epochs=600)
print(model.predict([10.0]))
print("Here is what I learned: {}".format(l0.get_weights()))
````

````
Weight, Bias = l0.get_weights()
W = f'{Weight[0][0]:.8}'
B = f'{Bias[0]:.8}'
#W, B
print(f"Thus, the learned relationship between X and Y was Y = {W}X", end='')
if float(B) > 0:
  print(f" + {B}")
else:
  print(f" - {B[1:]}")
````
