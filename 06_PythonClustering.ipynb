### **Clustering using scikit-learn <sup>[<a href="http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html" target="_blank">1</a>]</sup> (with pandas and matplotlib)**

```
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Load the Drive helper and mount
from google.colab import drive

# This will prompt for authorization.
drive.mount('/content/drive')
```

```
# Now, define your path of the input files
data_path = '/content/drive/My Drive/ML/'
```

```
# Read Cities.csv into dataframe
f = data_path + 'Cities.csv'
cities = pd.read_csv(f)
cities
```

```
# K-means clustering based on longitude-latitude
features = ['longitude', 'latitude']
k = 5
kmeans = KMeans(k)
kmeans.fit(cities[features])

labels = kmeans.labels_
# print(labels) #หมายเลขกลุ่ม
#print(labels)
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
# 1) Try to run several times and observe the clusters
# 2) Try different values for k
```

```
# Now with centroids
features = ['longitude', 'latitude']
k = 5
kmeans = KMeans(k)
kmeans.fit(cities[features])

labels = kmeans.labels_
centroids = kmeans.cluster_centers_  #mean
centroids.T
# centroids
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', c='black')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```

```
# k-means clustering based on temperature
features = ['temperature']
k = 3
kmeans = KMeans(k)
kmeans.fit(cities[features])

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(centroids)
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```
